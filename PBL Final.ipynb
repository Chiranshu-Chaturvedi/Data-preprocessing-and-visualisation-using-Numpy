```python
import numpy as np
import datetime as dt
import pandas as pd
kk=pd.read_csv("IMDb Movies.csv")
kk['Censor']=kk['Censor'].str.replace('Not Rated','Not available')
kk['Director']=kk['Director'].str.replace('Directors:','')
kk
```

```python
kk['main_genre']=kk['main_genre'].str.replace('Film-Noir','Dark film')
kk[kk['main_genre']=='Dark film'].tail(5)
```

```python
kk.loc[kk.duplicated()].head(10)
```

```python
kk=(kk.loc[kk.duplicated()].reset_index(drop=True).copy())
```

```python
kk.loc[kk.duplicated()].head(10)
```

```python
#group by with multiple keys(columns)
df=pd.DataFrame(kk)
print(df)
print(df.groupby(['Year','Rating']).groups)
```

```python
#iterating through groups
grouped=df.groupby('Rating')
for name,group in grouped:
    print(name)
    print(group)
```

```python
#select single group
grouped=df.groupby('Year')
(grouped.get_group(2009))
```

```python
#aggregation
grouped=df.groupby('Year')
print(grouped['Rating'].agg([np.mean,np.sum,np.std]))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
kk2 = (
    df['main_genre']
    .value_counts()
    .head()
    .plot(kind='bar', title='Top 5 Main Genres'))
kk2.set_ylabel('Count')
kk2.set_xlabel('Movie Genre')
plt.show()
```
